```python
import pandas as pd

data_sa = pd.read_csv("Data_SA.csv",index_col=False)

freq_tra = list(data_sa.groupby('dest cou').agg({'Seats': "sum"}).sort_values(by="Seats").reset_index()["dest cou"])

high = freq_tra[-5:]
low = freq_tra[0:5]
mid = freq_tra[20:25]
places = high + low + mid

data_freq_sa = data_sa[data_sa["dest cou"].isin(places)]

data_freq_sa.to_csv("Frequent from SA.csv")
```

```python
data_gb = pd.read_csv("Data_GB.csv",index_col=False)

freq_tra = list(data_gb.groupby('dest cou').agg({'Seats': "sum"}).sort_values(by="Seats").reset_index()["dest cou"])

high = freq_tra[-5:]
low = freq_tra[0:5]
mid = freq_tra[20:25]
places = high + low + mid

data_freq_gb = data_gb[data_gb["dest cou"].isin(places)]

data_freq_gb.to_csv("Frequent from GB.csv")
```
